#### Decision Tree
`판단 기준`
- **Information Gain** : 1-Entropy를 의미하며 데이터 집합의 혼잡도를 의미하는데, 서로 다른 값이 섞여 있으면 Entropy가 높아진다. 큰값을 가질수록 분류기로부터 얻는 정보가 많으므로 좋다고 할 수 있음.
- **Gini Index** : 1-sum(p^2), 불순도를 나타냄. 분류기가 영역1,2를 구분하고 각각의 영역에서 클래스1,2가 각각 얼마나 포함되어 있는지를 나타내 줌. 이 계수를 활용하면 트리에서 구분된 Box 내 같은 클래스가 얼마나 있는지를 판단 가능하다. 지니계수가 작을수록 분류를 잘 하고 있는 것이다. 참고로 CART 알고리즘은 Gini index 기준이다.

|장점|단점|
|:---:|:---:|
|쉽고 직관적이다.|Overfitting 발생|
|전처리 영향이 크지 않다.|Overfitting 극복하려면 모수 튜닝 필요|

- Decision Tree 는 특히, 데이터 스케일링이나 정규화 등 전처리의 영향이 적다.
