## DataFrame <-> List, dict, numpy(ndarray) 상호 변환

```python
### 1차원 데이터 변환

import numpy as np

# List를 이용해 데이터 프레임 생성 
col_name1= ['col1']
list1= [1,2,3]
df_list1=pd.DataFrame(list1, columns=col_name1)
print(df_list1)


# 넘파이 ndarray 이용해 데이터 프레임 생성 
array1 = np.array(list1)
print('array1 shape:', array1.shape)
df_list1=pd.DataFrame(array1, columns=col_name1)
print(df_list1)
```

```python
### 2차원 데이터 변환

col_name2=['col1', 'col2', 'col3']# 3개의 컬럼명이 필요함. 
list2 = [[1, 2, 3],
         [11, 12, 13]] # 2행x3열 형태의 리스트와 ndarray 생성 한 뒤 이를 DataFrame으로 변환. 

# List를 이용해 데이터 프레임 생성 
df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)

# 넘파이 ndarray 이용해 데이터 프레임 생성 
array2 = np.array(list2)
print('array2 shape:', array2.shape )
df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

```python
# dict -> 데이터프레임으로 변환하기

my_dict= {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict=pd.DataFrame(my_dict)
print(df_dict)
```

## DataFrame -> ndarray, List, Dict로 변환

```python
# dataframe -> ndarray로 변환.

array3 = df_dict.values # value값이 곧 ndarray형태이므로 잘 변환됨을 아래에서도 알 수 있음. 
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)

list3 = df_dict.values.tolist() # ndarray로 변환된 객체에 바로 tolist() 메서드로 list로 변환하기"
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

dict3 = df_dict.to_dict('list') # to_dict() 는 dict형태로 변환이 가능하도록 돕는데, 'list' 인자는 values 값의 타입을 의미한다. 
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

## Missing Data Handling
- NULL인 경우를 결손데이터라고 하고 넘파이의 NaN으로 표시한다. 머신러닝을 위해서는 대체해야 함. 
- isna() : 확인
- fillna() : 다른 값으로 대체

```python
titanic_df.isna().head(4)
```

```python
titanic_df.isna().sum()
```

- 대체하기 : fillna()

```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3) 

# Cabin에 있던 결측값이 C000으로 대체가 됨.
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna( titanic_df['Age'].mean() ) # 변경하지 않으려면 ,inplace= True 옵션 지정해주기
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')

titanic_df.isna().sum()
```

## apply, lambda 식으로 데이터 가공 
- lambda : 파이썬에서 함수형 프로그래밍을 지원하기 위해 만들어졌음.

```python
def get_square(a):
    return a**2

print( '3의 제곱은 :', get_square(3))
```

```python
lambda_square = lambda x : x**2
lambda_square(3)
```

```python
#list(map(lambda x: x**2, 3)) # 이렇게 뒤에 인자에 값 1개만 넣어주면 에러남.. 
list(map(lambda x: x**2,[1,2,3])) # 이렇게 뒤에 list값 넣어줘야 출력 잘 나옴.
```

- 그래서 이걸 pandas에 적용한다면, .apply(input : output) 이렇게 써야함.

```python
titanic_df['Name_len']= titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name','Name_len']].head(3)
```

```python
titanic_df['Child_Adult']= titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(5)
```

```python
# 조건 여러개 사용하고 싶으면 else() 활용해서 중첩 if 써야 함. 
# (elif 사용 안됨.)

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x:'Child' if x<15 else('Adult' if x<=60 else 'Elderly') )
titanic_df['Age_cat'].value_counts()
```

```python
# 조건이 많은 경우는 아래처럼 함수 생성하는 것이 나음.
# 나이에 따라 세분화된 분류를 수행하는 함수 생성. 
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    
    return cat

# lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정. 
# get_category(X)는 입력값으로 ‘Age’ 컬럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
```
